##IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
```

```python
import zipfile
```

##UNZIPPING THE DATASET

```python
unzip=zipfile.ZipFile("/content/Anonymize_Loan_Default_data.csv.zip")
unzip.extractall()
unzip.close()
```

```python
dataset=pd.read_csv("/content/Anonymize_Loan_Default_data.csv",encoding='latin1')
```

```python
dataset.term.value_counts()
```

##PREPROCESSING THE DATASET

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
le=LabelEncoder()
dataset["term"]=le.fit_transform(dataset["term"])
dataset["emp_length"]=le.fit_transform(dataset["emp_length"])
dataset["home_ownership"]=le.fit_transform(dataset["home_ownership"])
dataset["verification_status"]=le.fit_transform(dataset["verification_status"])
dataset["loan_status"]=le.fit_transform(dataset["loan_status"])
dataset["purpose"]=le.fit_transform(dataset["purpose"])
```

```python
dataset.columns
```

```python
dataset.head()
```

```python
dataset.drop(columns=["id",
                      "member_id",
                      "last_credit_pull_d",
                      "next_pymnt_d",
                      "last_pymnt_d",
                      "earliest_cr_line",
                      "addr_state",
                      "zip_code",
                      "issue_d",
                      "revol_util",
                      "mths_since_last_delinq"],inplace=True)
```

```python
l=[]
for i in dataset:
  print(dataset[i].isna().value_counts())
```

##FILLING THE MISSING VALUES

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
x=pd.DataFrame(imp.fit_transform(dataset))
```

```python
dataset.dtypes
```

```python
x
```

```python
y=x[25]
x.drop(columns=[0,25],inplace=True)
```

##SPLITTING THE DATASET INTO TEST AND TRAIN SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train),len(x_test)
```

```python
import lightgbm as lg
```

##CLASSIFING THE DATA USING LIGHTGBM

```python
lc=lg.LGBMClassifier()
lc.fit(x_train,y_train)
```

```python
y_pred_1=lc.predict(x_test)
y_pred_1
```

##EVALUTAING THE PEFORMANCE OF MODEL ON TEST SET

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
cm=confusion_matrix(y_test,y_pred_1)
print(cm)
ac=accuracy_score(y_test,y_pred_1)
print(ac)
f1=f1_score(y_test,y_pred_1)
print(f1)
```
